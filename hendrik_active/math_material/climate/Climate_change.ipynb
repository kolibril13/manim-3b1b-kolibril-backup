```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
plt.rcParams['figure.dpi'] = 150   
```

```python
x=pd.read_csv("Default Dataset.csv")
```

```python
x=x.sort_values("wavelength")
x["wavelength"]= x["wavelength"]*100
```

```python
plt.plot(x.wavelength, x.intensity)
plt.xlim(0,200000)
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=go.Scatter(
                x=x.wavelength,
                y=x.intensity,
                mode='markers',
                hovertext=x.index
))
fig.show()
```

```python
from scipy.constants import k, h, c ,pi
T=294

def freq_vals(k):
    return c*k

freqs= freq_vals(x.wavelength)
rad= 2* h * freqs**3/c**2 * 1/(np.exp( (h*freqs) / (k*T) )-1)
rad= rad/max(rad) *max(x.intensity)+0.15*max(x.intensity)
plt.plot(x.wavelength, rad)
plt.plot(x.wavelength, x.intensity)
plt.xlim(0,200000)
```
