```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
plt.rcParams['figure.dpi'] = 200  
```

```python
for i in range(0,150):
    plt.cla()

    # implementation 1
    plt.ylim(-105,105)
    x = np.linspace(-100,100,400)
    def g(x,k_size):
        k_all=np.linspace(0,0.5,k_size)
        x_m, k_m = np.meshgrid(x,k_all)
        matrix= np.exp(-1j*(k_m*x_m - (k_m/(k_m**1.1+1))*(0+i) ))
        my_sum=np.sum(matrix,axis=0)
        return my_sum
    plt.axhline(c="k")
    plt.axvline(c="k")
    plt.plot(x,np.real(g(x,k_size=100)));
    name=  f"temp/p{i:03}"
    plt.savefig(name)
```
