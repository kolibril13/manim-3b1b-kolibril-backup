```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# initial parameters:
N=10000
R= 100 # number of repetitions
np.random.seed(42)
```

```python
%%time
all_walks_list= []
for r in range(0,R):
    rands=np.random.randint(2, size=N)
    rands=np.where(rands==0, -1, rands) 
    x_traj_list= [0]
    for n in range(0,N):  # faster way # 
        x_traj_list.append(rands[n]+x_traj_list[n])
    
    all_walks_list.append(x_traj_list)
    

#x_traj= np.zeros(N+1)  #slow idea
#for n in range(1,N):
#    x_traj[n]=(rands[n]+x_traj[n-1])
```

```python
for element in all_walks_list[1:10]:
    plt.plot(element)
```

## Sum up distances

```python
distances_all = []
for element in all_walks_list:
    distances_all.append(element[-1])
```

```python
distances_squared= np.array(distances_all)**2
```

```python
sns.distplot(distances_all, kde=False, rug=True);
```

```python
sns.distplot(distances_squared,kde=False, rug=True);
```

```python
np.mean(distances_all), np.mean(distances_squared)
```

```python
D=  (2*N)/np.mean(distances_squared)
```
